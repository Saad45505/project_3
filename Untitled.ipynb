Exploratory Data Analysis on Diwali Sales

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('saad.csv', encoding='latin1')

```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
ax = sns.countplot(x= 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
bx= sns.barplot(x='Gender', y='Amount', data= sales_gen, hue= 'Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
ax= sns.countplot(data = df, x='Age Group', hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.barplot(x= 'Age Group', y= 'Amount', data= sales_age)
```

From the above graphs we analyzed that most of buyers are women age between 26-35

```python
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(data=sales_state, x='State', y='Orders')
plt.xticks(rotation=90)  # Optional: Rotate x-axis labels if needed
plt.show()
```
